```python
import sqlite3
#connect to sqlite database
conn=sqlite3.connect('weather_data.db')
cur=conn.cursor()
# Get the most frequent weather condition per city
cur.execute("SELECT city, description, COUNT(description) as count FROM weather_data GROUP BY city, description ORDER BY count DESC")
weather_conditions = cur.fetchall()

print("Most Frequent Weather Conditions per City:")
for row in weather_conditions:
    print(f"{row[0]}: {row[1]} ({row[2]} occurrences)")
```

```python
import sqlite3
#connect to sqlite database
conn=sqlite3.connect('weather_data.db')
cur=conn.cursor()
# Get the minimum and maximum temperatures per city
cur.execute("SELECT city, MIN(temperature), MAX(temperature) FROM weather_data GROUP BY city")
temp_extremes = cur.fetchall()

print("Temperature Extremes per City:")
for row in temp_extremes:
    print(f"{row[0]}: Min Temp = {row[1]:.2f}°C, Max Temp = {row[2]:.2f}°C")
```
